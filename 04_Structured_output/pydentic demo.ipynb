```python
from pydantic import BaseModel

class Student(BaseModel):
    name: str
    

new_student = {"name": "Priyanshu"}
student=Student(**new_student)
print(student)
```

```python
from pydantic import BaseModel
from typing import Optional

class Student(BaseModel):
    name: str='Priyanshu'
    age: Optional[int] = None
    

new_student = {"name": "Priyanshu",'age': 25}
student=Student(**new_student)
print(student)
```

```python
from pydantic import BaseModel,EmailStr
from typing import Optional

class Student(BaseModel):
    name: str='Priyanshu'
    age: Optional[int] = None
    email: Optional[EmailStr] = None
    

new_student = {"name": "Priyanshu",'age': 25, 'email': "priya@example.com"}
student=Student(**new_student)
print(student)
```
